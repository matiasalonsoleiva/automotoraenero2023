```python
import pandas as pd
import matplotlib.pyplot as plt

######RANKING DE VENTAS POR MARCA VEHICULOS HIBRIDOS 2021##############

# Crear un dataframe con los datos
datos = {'Marca': ['TOYOTA', 'FORD', 'SUBARU', 'LEXUS', 'KIA', 'HYUNDAI', 'HAVAL', 'BMW', 'Total'],
         'Ventas 2021': [1389, 120, 109, 134, 36, 5, 0, 3, 1796]}
df = pd.DataFrame(datos)

# Eliminar la fila Total temporalmente para el grfico
df_temp = df[df['Marca'] != 'Total']

# Crear el grfico de barras
fig, ax = plt.subplots()
ax.bar(df_temp['Marca'], df_temp['Ventas 2021'], color='green')

# Agregar etiquetas a cada barra
for i, v in enumerate(df_temp['Ventas 2021']):
    ax.text(i, v, str(v), ha='center', va='bottom', fontsize=10)

# Ajustar los ejes y los ttulos del grfico
ax.set_title('Ranking de ventas por marca - Vehculos hbridos 2021')
ax.set_xlabel('Marcas')
ax.set_ylabel('Ventas')
ax.set_ylim([0, max(df_temp['Ventas 2021']) + 100])

# Aadir la leyenda
ax.legend(['Ventas 2021'], loc='upper right')

# Aadir el total de ventas fuera del grfico
total_ventas = df.loc[df['Marca'] == 'Total', 'Ventas 2021'].values[0]
plt.text(len(df_temp), total_ventas, 'Total de ventas: ' + str(total_ventas), ha='right', va='center', fontsize=10)
plt.rc('font', family='Arial')

plt.xticks(rotation=40)
plt.show()






```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

######RANKING DE VENTAS POR MARCA VEHICULOS ELECTRICOS 2021##############


# Crear el dataframe con los datos
data = {
    'Marca': ['MAXUS', 'PEUGEOT', 'DS', 'HYUNDAI', 'NISSAN', 'BMW', 'MINI', 'MG', 'AUDI', 'VOLVO', 'PORSCHE', 'RENAULT', 'HIGER', 'JAC', 'DFSK', 'OTROS'],
    'Ventas': [267, 17, 2, 48, 24, 8, 41, 25, 5, 22, 18, 17, 3, 1, 1, 57]
}

df = pd.DataFrame(data)

# Ordenar los valores por ventas de mayor a menor
df = df.sort_values('Ventas', ascending=False)

# Crear la figura y el grfico de barras
fig, ax = plt.subplots(figsize=(10, 6))
sns.set_style("whitegrid")

# Aadir los datos al grfico
sns.barplot(x='Marca', y='Ventas', data=df, ax=ax, color='lightgreen')

# Aadir los valores exactos en el eje y
for i, v in enumerate(df['Ventas']):
    ax.text(i, v, str(v), ha='center', fontsize=10, fontname='Arial')

#Aadir leyenda 

# Aadir la leyenda
ax.legend(['Ventas 2021'], loc='upper right')

#Fondo blanco

ax.set_facecolor('none')
sns.set_style("whitegrid")


# Ajustar los ttulos y etiquetas
ax.set_xlabel('Marcas', fontsize=12, fontname='Arial')
ax.set_ylabel('Ventas', fontsize=12, fontname='Arial')
ax.set_title('Ranking de ventas por marca - Vehculos elctricos 2021', fontsize=14, fontname='Arial')

plt.xticks(rotation=40)

plt.grid(False)
# Mostrar la figura
plt.show()
```
